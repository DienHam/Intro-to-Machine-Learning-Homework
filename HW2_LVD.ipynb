```python
#Import các thư viện cần thiết

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')
import random
import math
```

```python
#Load dữ liệu từ file
newdf = pd.read_csv("batdongsan.com.vn.csv", dtype = {"month": "string" , "id" : "string"})
print("DONE")
```

Đưa ra các câu lệnh để thực hiện các yêu cầu sau:

1. Minh họa boxblot của các trường số như price, bedrooms, square, ...

* Gợi ý: sns.boxplot(data_field)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=newdf['price'])
plt.title('Boxplot of Price')
plt.show()
# Insight: Identify outliers in property prices, which are common in real estate due to luxury listings.
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=newdf['square'])
plt.title('Boxplot of Square Footage')
plt.show()
# Insight: Helps visualize the distribution of property sizes and detect unusually large plots.
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=newdf['bedrooms'])
plt.title('Boxplot of Bedrooms')
plt.show()
# Insight: Most properties likely fall within a specific range (e.g., 1-4); outliers may indicate villas or boarding houses.
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=newdf['bathrooms'])
plt.title('Boxplot of Bathrooms')
plt.show()
# Insight: Shows the typical ratio of bathrooms per property.
```

2. Cho đoạn code minh họa Biểu đồ thể hiện số bài đăng theo tháng của năm 2020. Hãy thực hiện các yêu cầu cho các câu hỏi dưới đây:

```python
df1 = newdf.groupby('month').size().reset_index(name='number of records')
n = df1['month'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df1['month'], df1['number of records'], color=c, width=.5)
for i, val in enumerate(df1['number of records'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticklabels(df1['month'], rotation=60, horizontalalignment= 'right')
plt.title("Biểu đồ thể hiện số bài đăng theo tháng của năm 2020", fontsize=22)
plt.ylabel('number of records')
plt.show()
```

* Biểu đồ thể hiện giá nhà đất trung bình theo tháng (VND/m2)

```python
# Calculate unit price if not already present
if 'unit_price' not in newdf.columns:
    newdf['unit_price'] = newdf['price'] / newdf['square']

avg_price_month = newdf.groupby('month')['unit_price'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_price_month, x='month', y='unit_price', marker='o')
plt.title('Average Unit Price (VND/m2) by Month')
plt.xticks(rotation=45)
plt.show()
# Insight: Visualizes market trends over time to identify peak pricing periods.
```

* Biểu đồ mật độ giá nhà của tháng 4 và tháng 12 (VND/m2)

```python
plt.figure(figsize=(12, 6))
sns.histplot(newdf[newdf['month'] == '4']['unit_price'], color='blue', label='April', kde=True, element='step')
sns.histplot(newdf[newdf['month'] == '12']['unit_price'], color='orange', label='December', kde=True, element='step')
plt.title('Price Density Distribution: April vs December')
plt.legend()
plt.show()
# Insight: Compares the price distribution shift between the beginning and end of the year.
```

* Số lượng bất động sản phân theo số lượng bedrooms

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=newdf, x='bedrooms')
plt.title('Number of Properties by Bedroom Count')
plt.show()
# Insight: Reveals which apartment/house configurations are most prevalent in the market.
```

* Giá bất động sản trung bình theo số lượng bedrooms

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=newdf, x='bedrooms', y='price', estimator=np.mean)
plt.title('Average Total Price by Bedroom Count')
plt.show()
# Insight: Validates the expected correlation between property size (bedrooms) and total cost.
```

* Giá nhà đất trung bình theo quận (VND/m2)

```python
plt.figure(figsize=(15, 6))
district_price = newdf.groupby('district')['unit_price'].mean().sort_values(ascending=False)
sns.barplot(x=district_price.index, y=district_price.values)
plt.title('Average Price (VND/m2) by District')
plt.xticks(rotation=90)
plt.show()
# Insight: Highlights the most expensive geographical areas.
```

* Giá nhà đất trung bình theo các phường quận Hoàn Kiếm (VND/m2)

```python
hoan_kiem_df = newdf[newdf['district'] == 'Quận Hoàn Kiếm']
ward_price = hoan_kiem_df.groupby('ward')['unit_price'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=ward_price.index, y=ward_price.values)
plt.title('Average Price (VND/m2) in Hoan Kiem Wards')
plt.xticks(rotation=45)
plt.show()
# Insight: Drills down into localized pricing in a premium district.
```

* Giá nhà đất trung bình theo nhà đầu tư (VND/m2)

```python
investor_price = newdf.groupby('investor')['unit_price'].mean().sort_values(ascending=False).head(20)
plt.figure(figsize=(15, 6))
sns.barplot(x=investor_price.index, y=investor_price.values)
plt.title('Top 20 Average Prices (VND/m2) by Investor')
plt.xticks(rotation=90)
plt.show()
# Insight: Identifies high-end developers based on property valuations.
```

* Giá nhà đất trung bình theo dự án (VND/m2)

```python
project_price = newdf.groupby('project')['unit_price'].mean().sort_values(ascending=False).head(20)
plt.figure(figsize=(15, 6))
sns.barplot(x=project_price.index, y=project_price.values)
plt.title('Top 20 Average Prices (VND/m2) by Project')
plt.xticks(rotation=90)
plt.show()
# Insight: Pins down specific real estate projects with the highest market value.
```

* Giá bất động sản phân bố theo top 10 nhà đầu tư triệu /m2

```python
top_investors = newdf.groupby('investor')['unit_price'].mean().nlargest(10).index
plt.figure(figsize=(12, 6))
sns.boxplot(data=newdf[newdf['investor'].isin(top_investors)], x='investor', y='unit_price')
plt.title('Price Distribution for Top 10 Investors')
plt.xticks(rotation=90)
plt.show()
# Insight: Shows the price consistency or variance within projects of premium developers.
```

* Giá bất động sản phân bố theo dự án

```python
top_projects = newdf['project'].value_counts().head(10).index
plt.figure(figsize=(12, 6))
sns.boxplot(data=newdf[newdf['project'].isin(top_projects)], x='project', y='unit_price')
plt.title('Price Distribution by Most Listed Projects')
plt.xticks(rotation=90)
plt.show()
# Insight: Compares price stability across the most popular developments.
```

* Giá nhà đất trung bình theo hướng nhà (VND/m2)

```python
direction_price = newdf.groupby('direction')['unit_price'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=direction_price.index, y=direction_price.values)
plt.title('Average Price (VND/m2) by Direction')
plt.show()
# Insight: Determines if specific orientations (like South or Southeast) command a premium.
```

4. Mật độ giá nhà trung bình theo quận
